```python
import java.sql.*;
import java.util.*;

public class OnlineReservationSystem {
    private static final String DB_URL = "jdbc:mysql://localhost:3306/reservation_db";
    private static final String DB_USER = "your_username";
    private static final String DB_PASSWORD = "your_password";

    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        Connection connection = null;
        Statement statement = null;
        ResultSet resultSet = null;

        try {
            // Establish connection to the database
            connection = DriverManager.getConnection(DB_URL, DB_USER, DB_PASSWORD);
            statement = connection.createStatement();

            // Main loop for user interaction
            while (true) {
                System.out.println("Choose an option:");
                System.out.println("1. Login");
                System.out.println("2. Exit");
                int choice = scanner.nextInt();

                switch (choice) {
                    case 1:
                        // Login form
                        System.out.print("Enter username: ");
                        String username = scanner.next();
                        System.out.print("Enter password: ");
                        String password = scanner.next();

                        // Verify login credentials
                        String query = "SELECT * FROM users WHERE username='" + username + "' AND password='" + password + "'";
                        resultSet = statement.executeQuery(query);

                        if (resultSet.next()) {
                            // Successful login
                            System.out.println("Login successful!");

                            // Display main menu
                            while (true) {
                                System.out.println("Choose an option:");
                                System.out.println("1. Reservation");
                                System.out.println("2. Cancellation");
                                System.out.println("3. Logout");
                                choice = scanner.nextInt();

                                switch (choice) {
                                    case 1:
                                        // Reservation form
                                        System.out.print("Enter your name: ");
                                        String name = scanner.next();
                                        System.out.print("Enter your contact number: ");
                                        String contact = scanner.next();
                                        System.out.print("Enter train number: ");
                                        int trainNumber = scanner.nextInt();
                                        System.out.print("Enter class type (AC/SL/GN): ");
                                        String classType = scanner.next();
                                        System.out.print("Enter date of journey (YYYY-MM-DD): ");
                                        String dateOfJourney = scanner.next();
                                        System.out.print("Enter from (place): ");
                                        String fromPlace = scanner.next();
                                        System.out.print("Enter to (destination): ");
                                        String toPlace = scanner.next();

                                        // Insert reservation details into the database
                                        query = "INSERT INTO reservations (name, contact, train_number, class_type, date_of_journey, from_place, to_place) VALUES ('" + name + "', '" + contact + "', " + trainNumber + ", '" + classType + "', '" + dateOfJourney + "', '" + fromPlace + "', '" + toPlace + "')";
                                        statement.executeUpdate(query);
                                        System.out.println("Reservation successful!");
                                        break;
                                    case 2:
                                        // Cancellation form
                                        System.out.print("Enter PNR number: ");
                                        int pnrNumber = scanner.nextInt();

                                        // Retrieve reservation details based on PNR number
                                        query = "SELECT * FROM reservations WHERE pnr_number=" + pnrNumber;
                                        resultSet = statement.executeQuery(query);

                                        if (resultSet.next()) {
                                            // Display reservation details
                                            System.out.println("Reservation details:");
                                            System.out.println("PNR Number: " + resultSet.getInt("pnr_number"));
                                            System.out.println("Name: " + resultSet.getString("name"));
                                            System.out.println("Contact: " + resultSet.getString("contact"));
                                            System.out.println("Train Number: " + resultSet.getInt("train_number"));
                                            System.out.println("Class Type: " + resultSet.getString("class_type"));
                                            System.out.println("Date of Journey: " + resultSet.getString("date_of_journey"));
                                            System.out.println("From: " + resultSet.getString("from_place"));
                                            System.out.println("To: " + resultSet.getString("to_place"));

                                            System.out.println("Confirm cancellation? (y/n)");
                                            String confirm = scanner.next();

                                            if (confirm.equalsIgnoreCase("y")) {
                                                // Delete the reservation from the database
                                                query = "DELETE FROM reservations WHERE pnr_number=" + pnrNumber;
                                                statement.executeUpdate(query);
                                                System.out.println("Cancellation successful!");
                                            } else {
                                                System.out.println("Cancellation canceled.");
                                            }
                                        } else {
                                            System.out.println("Invalid PNR number.");
                                        }
                                        break;
                                    case 3:
                                        // Logout
                                        System.out.println("Logging out...");
                                        break;
                                    default:
                                        System.out.println("Invalid choice.");
                                }

                                if (choice == 3) {
                                    break;
                                }
                            }
                        } else {
                            System.out.println("Invalid login credentials.");
                        }
                        break;
                    case 2:
                        // Exit the program
                        System.out.println("Exiting...");
                        System.exit(0);
                    default:
                        System.out.println("Invalid choice.");
                }
            }
        } catch (SQLException e) {
            e.printStackTrace();
        } finally {
            // Close resources
            try {
                if (resultSet != null) {
                    resultSet.close();
                }
                if (statement != null) {
                    statement.close();
                }
                if (connection != null) {
                    connection.close();
                }
            } catch (SQLException e) {
                e.printStackTrace();
            }
        }
    }
}
```
